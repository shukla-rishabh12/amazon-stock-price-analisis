q:Kaunse 3 saal me stock ka daily return sabse zyada volatile raha?
ans: positive me hi sabse jyada volatile raha h 1998 me 0.7% ka 2002 me ~0.6% and 2003 0.5% and agr baat kre neg fluctuation me to 1999 me ~  -0.23% and 2001 -0.4%

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Prepare empty dictionary to hold monthly jump counts for each year
monthly_jump_data = {}

# Step 2: Loop through years and collect monthly jump counts
for year in range(2017, 2026):
    df_year = df[df['Year'] == year].copy()
    df_year = df_year.sort_values('Date')
    df_year['prev_close'] = df_year['Close'].shift(1)
    df_year['price_jump'] = df_year['Close'] > df_year['prev_close']
    df_year['Month'] = df_year['Date'].dt.month

    jumps = df_year[df_year['price_jump']]
    monthly_counts = jumps.groupby('Month').size()

    # Store monthly jump counts for the year
    monthly_jump_data[year] = monthly_counts

# Step 3: Create a DataFrame from the dictionary
monthly_df = pd.DataFrame(monthly_jump_data).fillna(0).astype(int)

# Optional: Month number to name for better x-axis
month_map = {
    1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun",
    7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"
}
monthly_df.index = monthly_df.index.map(month_map)

# Step 4: Plot grouped bar chart
monthly_df.plot(kind='bar', figsize=(14, 7), colormap='tab10')

plt.title("Monthly Stock Price Jump Comparison (20172025)", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Jump Days", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Year", bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
